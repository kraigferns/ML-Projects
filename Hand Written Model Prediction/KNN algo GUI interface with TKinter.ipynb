```python
import tkinter as tk
from PIL import ImageTk,Image,ImageDraw
import numpy as np
import cv2 
import joblib # used to load the ML model using this library

win=tk.Tk(className="Hand Written Digits") #main window
model=joblib.load('KNN Hand written digits.sav')

w,h=500,500
count=0

def eventFunction(event): #the event argument holds x and y co-ords
    #print(event.x,event.y)
    x1=event.x-20
    x2=event.x+20
    y1=event.y-20
    y2=event.y+20
    
    canvas.create_oval((x1,y1,x2,y2),fill="black")
    imgdrw.ellipse((x1,y1,x2,y2),fill="white")
    
def save(): #Used to save the image that is drawn
    #print("Saved is pressed")
    global count
    imgarray=np.array(img)
    imgarray=cv2.resize(imgarray,(8,8))
    cv2.imwrite('data/'+str(count)+'.jpg',imgarray)
    count+=1

def exit():
    win.destroy()
    
def clear():
    global img,imgdrw #if you don't call it, it'll create a new local object
    canvas.delete('all') #Clear the entire the canvas 
    
    #this is used to re initialize the image object
    img=Image.new('RGB',(500,500),(0,0,0)) #(type,size,colour)
    imgdrw=ImageDraw.Draw(img)
    
def predict():
    
    #First, making the image compatible to the code
    imgarray=np.array(img)
    imgarray=cv2.cvtColor(imgarray,cv2.COLOR_RGB2GRAY)
    imgarray=cv2.resize(imgarray,(8,8))
    
    #Flattening
    imgarray=np.reshape(imgarray,(1,64))
    
    #In our model, our bnw image gives not upto 8-bit(256 values)
    #But, only maximum upto 4-bit(16 values), due to image compression
    #Hence, the range is from 0 to 16 (Convert 0-255 to 0-16)
    imgarray=(imgarray/255.0)*15.0
    
    #prediction
    result=model.predict(imgarray)
    label.config(text="Predicted Digit: "+str(result))
    #Updating and configuring the label status on the window
    
    
    
canvas=tk.Canvas(win,width=500,height=500,bg="white")
# 3 ways of placing canvas on the window: pack,grid,place
canvas.grid(row=0,column=0,columnspan=4) #columnspan to place the buttons properly at the bottom


#Buttons
svbutton=tk.Button(win,text="Save",bg="light sea green", 
                   width=10,fg="White",font="arial 12 bold"
                   ,command=save)
svbutton.grid(row=1,column=0)

prbutton=tk.Button(win,text="Predict",bg="deep sky blue",
                   width=10,fg="White",font="arial 12 bold"
                  ,command=predict)
prbutton.grid(row=1,column=1)

clrbutton=tk.Button(win,text="Clear",bg="navy",fg="White",
                    width=10,font="arial 12 bold"
                   ,command=clear)
clrbutton.grid(row=1,column=2)

exbutton=tk.Button(win,text="Exit",bg="dark slate blue",
                   width=10,fg="White",font="arial 12 bold"
                  ,command=exit)
exbutton.grid(row=1,column=3)


label=tk.Label(win,text="Predicted Digit: None",bg="black"
               ,fg="white",font="arial 20 bold")
label.grid(row=2,column=0,columnspan=4)

canvas.bind('<B1-Motion>',eventFunction) 



# IMAGE OBJECT PART
img=Image.new('RGB',(500,500),(0,0,0)) #(type,size,colour)
imgdrw=ImageDraw.Draw(img)

#b1,b2,b3motion means selecting and holding the left key,middle key, rigght key on canvas
# another example - canvas.bind('<ButtonPress>',eventFunction)
win.mainloop()
```
